# Probabilidade 

Este capítulo tem uma visão geométrica da teoria da probabilidade e a relaciona com
conceitos familiares em álgebra linear e geometria. Essa abordagem conecta seu
intuição geométrica natural para as principais abstrações em probabilidade que podem
ajudar a orientar o seu raciocínio. Isso é particularmente importante na probabilidade
porque é fácil ser enganado. Precisamos de um pouco de rigor e alguma
intuição para nos guiar.

Na escola primária, você foi apresentado aos números naturais (ou seja, `1,2,3,..`)
e você aprendeu a manipulá-los por meio de operações como adição,
subtração e multiplicação. Mais tarde, você foi apresentado ao positivo e
números negativos e foram novamente ensinados a manipulá-los. No final das contas, você
foram introduzidos ao cálculo da reta real e aprenderam a
diferenciar, tomar limites, e assim por diante. Essa progressão proporcionou mais
abstrações, mas também ampliou o campo de problemas que você poderia resolver com sucesso
enfrentar. O mesmo se aplica à probabilidade. **Uma maneira de pensar sobre probabilidade é
como um novo conceito de número que permite resolver problemas que têm um
tipo de *incerteza* construída neles. Assim, a ideia-chave é que há algum
número, digamos $x$, com um companheiro de viagem, digamos, $f(x)$, e este companheiro
representa as incertezas sobre o valor de $x$ como se estivesse olhando para o número
$x$ através de uma janela congelada. O grau de opacidade da janela é
representado por $f(x)$.** Se quisermos manipular $x$, então temos que descobrir
o que fazer com $f(x)$. Por exemplo, se queremos $y= 2 x $, então temos que
entenda como $f(x)$ gera $f(y)$.

Onde está a parte *aleatória*? Para conceituar isso, precisamos ainda **outra
analogia: pense em uma colméia com o enxame ao seu redor representando $f(x)$,
e a própria colméia, que você mal consegue ver através do enxame, como $x$.**
peça aleatória é que você não sabe *qual* abelha em particular vai picar você!
Quando isso acontece, a incerteza evapora.
Até que isso aconteça, tudo o que temos é um conceito de enxame (ou seja, densidade de
abelhas) que representa uma *potencialidade* da qual a abelha acabará por picar.
Em resumo, uma maneira de pensar sobre probabilidade é como uma maneira de realizar
raciocínio matemático (por exemplo, adicionar, subtrair, tomar
limites) com uma noção de potencialidade que é assim transformada por estes
operações.

## Compreendendo a densidade de probabilidade

Para entender o coração da probabilidade moderna, que é construída
na **teoria de integração de Lesbesgue**, precisamos estender o conceito
de integração do cálculo básico. Para começar, vamos considerar o
seguinte função por partes

$$
f(x)  =  \begin{cases}
                1 & \mbox{if }  0 < x \leq 1 \\\
                2 & \mbox{if }  1 < x \leq 2 \\\
                0 & \mbox{otherwise }
            \end{cases}
$$

conforme mostrado na [Figure]. Em cálculo, **você aprendeu
Integração de Riemann**, que você pode aplicar aqui como



$$
\int_0^2 f(x) dx = 1 + 2 = 3
$$

que tem a **interpretação usual como a área dos dois retângulos
que compõem $f(x)$**. Até agora tudo bem.

Com a integração **Lesbesgue, a ideia é muito semelhante, exceto que
concentre-se no eixo y em vez de se mover ao longo do eixo x.**

A questão é dada $f(x) = 1$, qual é o conjunto de valores $x$ para o qual isso é
verdadeiro? Para o nosso exemplo, isso é verdadeiro sempre que $x\in (0,1]$.

Então agora temos uma correspondência entre os valores da função (ou seja, `1`
e `2`) e os conjuntos de valores $x$ para os quais isso é verdade, ou seja,
$\lbrace (0,1] \rbrace$ e $\lbrace (1,2] \rbrace$, respectivamente.

Para calcular a integral, simplesmente pegamos os valores da função (ou seja, `1,2`)
e **alguma forma de medir o tamanho do intervalo correspondente**
(ou seja, $\mu$) como no seguinte:

$$
\int_0^2 f d\mu = 1 \mu(\lbrace (0,1] \rbrace) + 2 \mu(\lbrace (1,2] \rbrace)
$$

Suprimimos parte da notação acima para enfatizar a generalidade.

**Observação
que obtemos o mesmo valor da integral como no caso de Riemann quando
$\mu((0,1]) = \mu((1,2]) = 1$.**

Ao introduzir a função $\mu$ como uma forma de
medindo os intervalos acima, introduzimos outro grau de liberdade em
nossa integração.

Isso acomoda muitas funções estranhas que não são tratáveis
usando a teoria usual de Riemann, mas nós o encaminhamos para uma introdução adequada a
Integração de Lesbesgue para estudo adicional [[jones2001lebesgue]](#jones2001lebesgue).

**Apesar disso,
o passo chave na discussão acima é a introdução da função $\mu$,
que encontraremos novamente como a chamada função de densidade de probabilidade.**

## Variáveis ​​aleatórias

A maioria das introduções à probabilidade vai direto para *variáveis ​​aleatórias* e
em seguida, explique como calcular integrais complicadas. O problema com isso
abordagem é que ela pula algumas das sutilezas importantes que vamos agora
considerar.

Infelizmente, o termo *variável aleatória* não é muito descritivo. UMA
melhor termo é *função mensurável*.


Para entender por que esse é um termo melhor,
temos que mergulhar nas construções formais de probabilidade por meio de um simples
exemplo.

Considere jogar um dado justo de seis faces. Existem apenas seis resultados possíveis,

$$
\Omega=\lbrace 1,2,3,4,5,6 \rbrace
$$

Como sabemos, se o dado for justo, então a probabilidade de cada resultado é $1/6$.

Para dizer isso formalmente, a medida de cada conjunto (ou seja, $\lbrace 1 \rbrace,\lbrace
2 \rbrace,\ldots,\lbrace 6 \rbrace$) é $\mu(\lbrace 1 \rbrace ) =\mu(\lbrace 2
\rbrace ) \ldots = \mu(\lbrace 6 \rbrace ) = 1/6$.

Neste caso, o $\mu$
A função que discutimos anteriormente é a função de massa de *probabilidade* usual, denotada por
$\mathbb{P}$. A função mensurável mapeia um conjunto em um
número na linha real. Por exemplo, $ \lbrace 1 \rbrace \mapsto 1 $ é
uma dessas funções desinteressantes.

Agora, é aqui que as coisas ficam interessantes.


**Suponha que você foi solicitado a construir uma moeda honesta a partir do dado justo.** Em outras palavras, queremos jogar o dado e registrar os resultados como se tivéssemos acabado de jogar uma moeda honesta. Como poderíamos fazer isso?


Uma maneira seria definir uma função mensurável que diga se o dado aparecer
`3` ou menos, então declaramos *cara* e caso contrário declaramos *coroa*. Este tem
alguma intuição forte por trás disso, mas vamos articulá-lo em termos de formal
teoria.


Esta estratégia cria dois conjuntos diferentes não sobrepostos $\lbrace
1,2,3 \rbrace$ e $\lbrace 4,5,6 \rbrace$. Cada conjunto tem a mesma probabilidade
*a medida*,

$$
\begin{eqnarray*}
\mathbb{P}(\lbrace 1,2,3 \rbrace)  & = &  1/2  \\\
\mathbb{P}(\lbrace 4,5,6 \rbrace) & = & 1/2
\end{eqnarray*}
$$

E o problema está resolvido. Toda vez que o dado sai
$\lbrace 1,2,3 \rbrace$, caso contrário, registramos cara e coroa.

Esta é a única maneira de construir um experimento de moeda honesta a partir de uma
morte justa?

Alternativamente, podemos definir os conjuntos como $\lbrace 1 \rbrace$,
$\lbrace 2 \rbrace$, $\lbrace 3,4,5,6 \rbrace$. Se definirmos o correspondente
medida para cada conjunto como o seguinte

$$
\begin{eqnarray*}
\mathbb{P}(\lbrace 1 \rbrace)  & = &  1/2  \\\
\mathbb{P}(\lbrace 2 \rbrace) & = & 1/2 \\\
\mathbb{P}(\lbrace 3,4,5,6 \rbrace) & = & 0
\end{eqnarray*}
$$

Então, temos outra solução para o problema da moeda honesta. Para
implementar isso, tudo o que fazemos é ignorar toda vez que o dado mostra '3,4,5,6' e
jogar novamente.

Isso é um desperdício, mas resolve o problema. Apesar disso,
esperamos que você possa ver como as partes interligadas da teoria fornecem uma
estrutura para transportar a noção de incerteza/potencialidade de um problema
para o próximo (por exemplo, do dado justo para a moeda justa).

**Vamos considerar um problema um pouco mais interessante em que jogamos dois dados.**

Nós
assume que cada lance é *independente*, o que significa que o resultado de um não
não influenciar o outro. Quais são os conjuntos neste caso? são todos pares
de resultados possíveis de dois lances, como mostrado abaixo,

$$
\Omega = \lbrace (1,1),(1,2),\ldots,(5,6),(6,6) \rbrace
$$

Quais são as medidas de cada um desses conjuntos?
  
  Em virtude do
reivindicação de independência, a medida de cada um é o produto das respectivas medidas
de cada elemento. Por exemplo,

$$
\mathbb{P}((1,2)) = \mathbb{P}(\lbrace 1 \rbrace) \mathbb{P}(\lbrace 2 \rbrace) = \frac{1}{6^2}
$$

Com tudo isso estabelecido, podemos perguntar o seguinte
pergunta:

**qual é a probabilidade de que a soma dos dados seja igual a
Sete?**

Como antes, a primeira coisa a fazer é caracterizar o
função mensurável para isso como $X:(a,b) \mapsto (a+b)$. Em seguida nós
associe todos os pares $(a,b)$ à sua soma. Podemos criar um
Dicionário Python para isso, como mostrado,

```python
d={(i,j):i+j for i in range(1,7) for j in range(1,7)}
```

```python
d = dict(d) # python 3
```

O próximo passo é coletar todos os pares $(a,b)$ que somam
cada um dos valores possíveis de dois a doze.

```python
from collections import defaultdict
dinv = defaultdict(list)
```

```python
for i,j in iter(d.items()):
    print([j,i])
```

```python
for i,j in iter(d.items()): # python 3
    dinv[j].append(i)
```

```python
dinv
```

O objeto `defaultdict` do módulo interno de coleções cria dicionários com

O próximo passo é calcular a probabilidade medida para cada um desses itens.
Usando a suposição de independência, isso significa que temos que calcular a soma dos
produtos das probabilidades de itens individuais em `dinv`.

Porque nós sabemos que
cada resultado é igualmente provável, cada termo na soma é igual a $1/36$.

Assim, todos
temos que fazer é contar o número de itens na lista correspondente para cada
chave em `dinv` e divida por `36`.

Por exemplo, `dinv[11]` contém `[(5, 6),
(6, 5)]`. A probabilidade de `5+6=6+5=11` é a probabilidade deste conjunto que
é composto pela soma das probabilidades dos elementos individuais
`{(5,6),(6,5)}`.

Neste caso, temos $\mathbb{P}(11) = \mathbb{P}(\lbrace
(5,6) \rbrace)+ \mathbb{P}(\lbrace (6,5) \rbrace) = 1/36 + 1/36 = 2/36$.
Repetindo este procedimento para todos os elementos, derivamos a massa de probabilidade
funcionar como mostrado abaixo,

```python
iter(dinv.items())
```

```python
X=[[i,len(j)/36.] for i,j in iter(dinv.items())]
```

## Variáveis ​​Aleatórias Contínuas

**As mesmas ideias funcionam com variáveis ​​contínuas, mas gerenciando os conjuntos
torna-se mais complicada porque a linha real, ao contrário dos conjuntos discretos, tem muitos
propriedades limitantes já incorporadas a ele que precisam ser tratadas
com cuidado.**

No entanto, vamos começar com um exemplo que deve
ilustrar as ideias análogas.

Suponha que uma variável aleatória $X$ seja
uniformemente distribuído no intervalo unitário. Qual é a probabilidade
que a variável assume valores menores que 1/2?

Para construir a intuição no caso discreto, vamos voltar ao nosso
experimento de lançamento de dados com o dado justo. A **soma dos valores dos dados
é uma função mensurável,**

$$
Y \colon \lbrace 1,2,\dots,6 \rbrace^2 \mapsto \lbrace 2,3,\ldots, 12 \rbrace
$$

**Ou seja, $Y$ é um mapeamento do produto cartesiano de conjuntos para um
conjunto discreto de resultados.**

Para calcular as probabilidades do conjunto de
resultados, precisamos derivar a medida de probabilidade para $Y$, $\mathbb{P}_Y$,
das medidas de probabilidade correspondentes para cada dado. Nossa discussão anterior
passou pela mecânica disso. Isso significa que

$$
\mathbb{P}_Y \colon \lbrace 2,3,\ldots,12 \rbrace \mapsto [0,1]
$$

Observe que há uma separação entre a definição da função e onde o
itens-alvo da função são medidos em probabilidade.

Mais francamente,

$$
Y \colon A \mapsto B
$$

Com,

$$
\mathbb{P}_Y \colon B \mapsto [0,1]
$$

**Assim, para calcular $\mathbb{P}_Y$, que é derivado
de outras variáveis ​​aleatórias, temos que expressar as classes de equivalência
em $B$ em termos de seus conjuntos de progenitor $A$.**

A situação para variáveis ​​contínuas segue o mesmo padrão, mas
com muitos detalhes técnicos mais profundos que vamos pular. Para o contínuo
caso, a variável aleatória é agora,

$$
X \colon \mathbb{R} \mapsto \mathbb{R}
$$

com medida de probabilidade correspondente,

$$
\mathbb{P}_X \colon \mathbb{R} \mapsto [0,1]
$$

**Mas onde estão os conjuntos correspondentes aqui? Tecnicamente, estes são os
Conjuntos *Borel*, mas podemos apenas pensar neles como intervalos.**

Voltando ao nosso
pergunta, qual é a probabilidade de que uma variável aleatória uniformemente distribuída
no intervalo da unidade leva valores menores que $1/2$?

Reformulando esta pergunta
de acordo com o quadro, temos o seguinte:

$$
X \colon [0,1] \mapsto [0,1]
$$

com correspondente,

$$
\mathbb{P}_X \colon [0,1] \mapsto [0,1]
$$

**Para responder à pergunta, pela definição do aleatório uniforme
variável no intervalo da unidade, calculamos a seguinte integral,**

$$
\mathbb{P}_X([0,1/2]) = \mathbb{P}_X(0 < X < 1/2) =  \int_0^{1/2} dx  = 1/2
$$

onde $dx$ da integral acima varre intervalos do
Tipo $B$. A medida de qualquer intervalo $dx$ (ou seja, conjunto do tipo $A$) é igual a
$dx$, por definição da variável aleatória uniforme. Para obter todas as partes móveis
em uma integral rica em notação, também podemos escrever isso como,

$$
\mathbb{P}_X(0 < X < 1/2) = \int_0^{ 1/2 } d\mathbb{P}_X(dx) = 1/2
$$

**Agora, vamos considerar um exemplo um pouco mais complicado e interessante. Como
antes, suponha que temos uma variável aleatória uniforme, $X$ e vamos introduzir
outra variável aleatória definida,**

$$
Y = 2 X
$$

Agora, **qual é a probabilidade de $0 < Y < \frac{1}{2}$?**


Para expressar isso em nossa estrutura, escrevemos,

$$
Y \colon [0,1] \mapsto [0,2]
$$

com correspondente,

$$
\mathbb{P}_Y \colon [0,2] \mapsto [0,1]
$$

Para responder à questão, precisamos medir o conjunto $[0,1/2]$, com
a medida de probabilidade para $Y$, $\mathbb{P}_Y([0,1/2])$.

Como podemos fazer isso?

Como $Y$ é derivado da variável aleatória $X$, como acontece com o dado justo
experimento de lançamento, temos que criar um conjunto de equivalências no alvo
espaço (ou seja, conjuntos do tipo $B$) que refletem de volta no espaço de entrada (ou seja,
conjuntos do tipo $A$).

**Ou seja, qual é o intervalo $[0,1/2]$ equivalente em termos
da variável aleatória $X$?**

Porque, funcionalmente, $Y=2 X$, então o tipo $B$
intervalo $[0,1/2]$ corresponde ao intervalo $[0,1/4]$ do tipo $A$.

De
medida de probabilidade de $X$, calculamos isso com a integral,

$$
\mathbb{P}_Y([0,1/2]) =\mathbb{P}_X([0,1/4])=  \int_0^{1/4} dx  = 1/4
$$

Now, **let's up the ante and consider the following random variable,**

$$
Y = X^2
$$

onde agora $X$ ainda está distribuído uniformemente, mas agora sobre o
intervalo $[-1/2,1/2]$.

Podemos expressar isso em nossa estrutura como,

$$
Y \colon [-1/2,1/2] \mapsto [0,1/4]
$$

Com correspondente,

$$
\mathbb{P}_Y \colon [0,1/4] \mapsto [0,1]
$$

Quanto é $\mathbb{P}_Y(Y < 1/8)$? Em outras palavras, qual é o
medida do conjunto $B_Y= [0,1/8]$?


Como antes, porque $X$ é derivado de nosso
variável aleatória uniformemente distribuída, temos que refletir o conjunto $B_Y$ em
conjuntos do tipo $A$.


**A única coisa a reconhecer é que porque $X^2$
é simétrico em relação a zero, todos os conjuntos $B_Y$ refletem de volta em dois conjuntos.**


**Isso significa que para qualquer conjunto $B_Y$, temos a correspondência $B_Y = A_X^+ \cup
A_X^{-}$. Então nós temos,**

$$
B_Y=\Big\lbrace 0<Y<\frac{1}{8}\Big\rbrace=\Big\lbrace 0<X<\frac{1}{\sqrt{8}} \Big\rbrace \bigcup \Big\lbrace -\frac{1}{\sqrt {8}}<X<0 \Big\rbrace
$$

A partir desta perspectiva, temos a seguinte solução,

$$
\mathbb{P}_Y(B_Y)=\mathbb{P}(A_X^+)/2 + \mathbb{P}(A_X^{-})/2
$$

onde o $\frac{1}{2}$ vem da normalização do $\mathbb{P}_Y$ para 1. Também,

$$
A_X^+    = \Big\lbrace 0<  X<\frac{1}{\sqrt{8}}   \Big\rbrace
$$

$$
\
A_X^{-}  = \Big\lbrace -\frac{1}{\sqrt {8}} <  X<0 \Big\rbrace
$$

Portanto,

$$
\mathbb{P}_Y(B_Y) = \frac{1}{2\sqrt 8} + \frac{1}{2\sqrt 8}
$$

porque $\mathbb{P}(A_X^+) =\mathbb{P}(A_X^-) = 1/\sqrt 8$.
  
  
  vamos
veja se isso sai usando o método usual de transformação de variáveis ​​de
cálculo. Usando este método, a densidade

$f_Y(y) dy = f_X(x)dx$,

$f_Y(y) = f_X(x)\frac{dx}{dy}\big|_{x\para y}$,

$y = x^2$

$x = y^{1/2}$

$\frac{dx}{dy} = \frac{1}{2 y^{1/2}}$

$f_Y(y) = f_X(\sqrt y)/(2 \sqrt y) =
\frac{1}{2 \sqrt y} $. Então, obtemos,

$$
\int_0^{\frac{1}{8}} \frac{1}{2 \sqrt y} dy = \frac{1}{\sqrt 8}
$$

que é o que obtivemos usando o método sets.

## Transformação de Variáveis ​​Além do Cálculo

Suponha que $X$ e $Y$ estejam uniformemente distribuídos no intervalo unitário e
defina $Z$ como

$$
Z = \frac{X}{Y-X}
$$

Qual é o $f_Z(z)$? Se você tentar isso usando o cálculo usual
método, você falhará (experimente!).

O problema é que um dos pré-requisitos técnicos para o método de cálculo não está em vigor.

A principal observação é que $Z \notin [-1,0]$.

**Ambos $X$ e $Y$ são positivos.**

- **Assumindo $Y>X$, então $Y-X>0$, e então $Z>0$;**
- **Assumindo $Y<X$, então $Y-X<0$, e então $Z<-1$.**

Se isso fosse possível, os $X$ e $Y$ teriam sinais diferentes, o que não pode acontecer,
dado que $X$ e $Y$ são uniformemente distribuídos por $[0,1]$.

Agora, vamos considerar quando $Z>0$. Neste caso, $Y>X$ porque $Z$ não pode ser positivo
por outro lado.

Para a função de densidade, estamos interessados ​​no conjunto $\lbrace 0 < Z < z \rbrace $. queremos calcular

$$
\mathbb{P}(Z<z) = \int \int B_1 dX dY
$$

Com,

$$
B_1 = \lbrace 0 < Z < z \rbrace
$$

Agora, temos que traduzir esse intervalo em um intervalo
relevantes para $X$ e $Y$.

- Para $0 < Z$, temos $ Y > X$.

- Para $Z < z $, temos $\frac{X}{Y-X}<z \implies
Y > X(1/z+1)$.

Juntando isso dá

$$
A_1 = \lbrace \max (X,X(1/z+1)) < Y < 1 \rbrace
$$

Integrando isso sobre $Y$ como segue,

$$
\int_0^1\lbrace\max(X,X(1/z+1))<Y<1 \rbrace dY = 1 - \max(X,X(1/z+1)) =\\
1 -X(1/z+1) =\frac{z-X-Xz}{z}\\\mbox{ where } z > \frac{X}{1-X}
$$

Note que

$\max(X,X(1/z+1))$ com $z=\frac{X}{1-X}$ é $\max(X,X)=X$.

caso contrário $\max(X,X(1/z+1)) = X(1/z+1)$ para $z > \frac{X}{1-X}$.

Resolvendo para $X$, temos

$ X < \frac{Z}{1+Z}$,

e integrando isso mais uma vez sobre $X$ dá

$$
\int_0^{\frac{z}{1+z}} \frac{-X+z-Xz}{z} dX = \frac{z}{2(z+1)} \mbox{ where }  z > 0
$$

Observe que este é o cálculo para a *probabilidade*
em si, não a função de densidade de probabilidade. Para conseguir isso, tudo o que temos
a fazer é diferenciar $\left(f_Z(z)=\frac{d P(Z<z)}{dz}\right)$ a última expressão a obter

$$
f_Z(z) = \frac{1}{(z+1)^2} \mbox{ where }  z > 0
$$

Agora precisamos calcular essa densidade usando o mesmo processo
para quando $Z < -1$.

Queremos o intervalo $ Z < z $ para quando $Z < -1$.

Para um $z$ fixo, isso é equivalente a $\frac{X}{Y-X}<z \implies X(1+1/z) < Y$.

Como $z$ é negativo (e $z<-1$), isso também significa que

$(1+1/z) > 0 \implies X < Y \implies Y > X$.

Nestes termos, podemos
tem a seguinte integral,

$$
\int_0^1 \lbrace X(1/z+1) <Y< X\rbrace dY = X - X(1/z+1)\\
-\frac{X}{z} \mbox{ where }  z < -1
$$

e integrando isso mais uma vez sobre $X$ dá o seguinte

$$
-\frac{1}{2 z} \mbox{ where }  z < -1
$$

Para obter a densidade para $z<-1$, diferenciamos isso com
em relação a $z$ para obter o seguinte,

$$
f_Z(z) = \frac{1}{2 z^2} \mbox{ where }  z < -1
$$

Juntando tudo isso, obtemos,

$$
f_Z(z) = 
\begin{cases}
      \frac{1}{(z+1)^2}     & \mbox{if }  z > 0 \\\
      \frac{1}{2 z^2}       & \mbox{if }  z < -1 \\\
      0 & \mbox{otherwise }
\end{cases}
$$

integra-se a um. Mas vamos tentar simular esse processo.

## Variáveis ​​Aleatórias Independentes

A independência é uma suposição padrão. Matematicamente, o
condição necessária e suficiente para a independência entre dois
variáveis ​​aleatórias $X$ e $Y$ é o seguinte:

$$
\mathbb{P}(X,Y) = \mathbb{P}(X)\mathbb{P}(Y)
$$

Duas variáveis ​​aleatórias $X$ e $Y$
são *não correlacionados* se,

$$
\mathbb{E}[(X-\overline{X})(Y-\overline{Y})]=0
$$

onde $\overline{X}=\mathbb{E}(X)$. Uma condição semelhante é que $X$ e $Y$ não são correlacionados quando seu coeficiente de correlação é zero:
 
$$\rho(X,Y) = \frac{Cov[X,Y]}{\sqrt{Var[X]Var[Y]}} = 0
$$

onde $Cov[X,Y] = \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y]$ é a convariância entre $X$ e $Y$.

Observe que $Cov[X,Y] = \mathbb{E}[(X-\overline{X})(Y-\overline{Y})]$.

Observe também que $Cov[X,Y] = 0$ implica $\mathbb{E}[XY] = \mathbb{E}[X]\mathbb{E}[Y]$, de modo que independência implica não correlação.

**Variáveis ​​aleatórias não correlacionadas às vezes são chamadas de variáveis ​​aleatórias *ortogonais*.**

**A não correlação é uma propriedade mais fraca do que a independência, no entanto**. Por exemplo, considere o
variáveis ​​aleatórias discretas $X$ e $Y$ distribuídas uniformemente sobre o conjunto
$\lbrace 1,2,3 \rbrace$ onde

$$
X =  
\begin{cases} 
1 & \mbox{if } \omega =1 \\\
0 & \mbox{if } \omega =2  \\\
-1 & \mbox{if } \omega =3
\end{cases}
$$

e também,

$$
Y =  
\begin{cases} 
0 & \mbox{if } \omega =1 \\\
1 & \mbox{if } \omega =2 \\\
0 & \mbox{if } \omega =3
\end{cases}
$$

Assim, $\mathbb{E}(X)=0$ e $\mathbb{E}(X Y)=0$, então $X$ e $Y$ não são correlacionados. No entanto, temos

$$
\mathbb{P}(X=1,Y=1)=0\neq \mathbb{P}(X=1)\mathbb{P}(Y=1)=\frac{1}{9}
$$

Portanto, essas duas variáveis ​​aleatórias *não* são independentes.
**Assim, não correlação não implica independência, geralmente.**

Mas
existe o importante caso de variáveis ​​aleatórias gaussianas para as quais
sim. Para ver isso, considere a função de densidade de probabilidade
para duas variáveis ​​aleatórias Gaussianas de média zero e variância unitária $X$ e
$S$,

$$
f_{X,Y}(x,y) = \frac{e^{\frac{x^2-2 \rho x
   y+y^2}{2 \left(\rho^2-1\right)}}}{2 \pi 
   \sqrt{1-\rho^2}}
$$

```python
def plot_dg(rho=1e-16):
    x, y = np.meshgrid(np.linspace(-3,3),np.linspace(-3,3))
    z = np.exp((x**2 + y**2 - 2*rho*x*y)/(2*(rho**2-1)))/(2*np.pi*np.sqrt(1-rho**2))
    f, ax = plt.subplots()
    ax.contourf(x,y,z,cmap='viridis')
    ax.set_aspect('equal')
```

onde $\rho:=\mathbb{E}(X Y)$ é o coeficiente de correlação. Dentro
o caso não correlacionado onde $\rho=0$, os fatores da função de densidade de probabilidade
no seguinte,

$$
f_{X,Y}(x,y)=\frac{e^{-\frac{1}{2}\left(x^2+y^2\right)}}{2\pi}=\frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}}\frac{e^{-\frac{y^2}{2}}}{\sqrt{2\pi}}    =f_X(x)f_Y(y)
$$

o que significa que $X$ e $Y$ são independentes.

Independência e independência condicional estão intimamente relacionadas, como no seguinte:

$$
\mathbb{P}(X,Y\vert Z) =\mathbb{P}(X\vert Z) \mathbb{P}(Y\vert Z)
$$

que diz que $X$ e $Y$ e condicionado independente

**O condicionamento de variáveis ​​aleatórias independentes pode quebrar
sua independência.**

Apenas lembrando que a probabilidade condicional é:

$$P(A|B) = \frac{P(A,B)}{P(B)}$$

onde $P(A,B) = P(A\cap B)$ é a probabilidade de $A$ e $B$ acontecerem.

Por exemplo, considere duas variáveis ​​aleatórias independentes distribuídas por Bernoulli, $X_1, X_2\in\lbrace 0,1
\rbrace$. Definimos $Z=X_1+X_2$. Note que $Z\in \lbrace
0,1,2 \rbrace$. No caso em que $Z=1$, temos,

$$
\mathbb{P}(X_1\vert Z=1) >0
$$

$$
\
\mathbb{P}(X_2\vert Z=1) >0
$$

Embora $X_1,X_2$ sejam independentes,
após o condicionamento em $Z$, temos o seguinte,

$$
\mathbb{P}(X_1=1,X_2=1\vert Z=1)=0\neq \mathbb{P}(X_1=1\vert Z=1)\mathbb{P}(X_2=1\vert Z=1)
$$

Assim, condicionar em $Z$ quebra a independência de
$X_1,X_2$.

**Isso também funciona na direção oposta ---
o condicionamento pode tornar as variáveis ​​aleatórias dependentes independentes.**


Defina $Z_n=\sum_i^n X_i$ com $X_i$ independente, valor inteiro
variáveis ​​aleatórias.

As variáveis ​​$Z_n$ são dependentes porque empilham o mesmo conjunto telescópico de variáveis ​​$X_i$. Considere o seguinte,

<!-- Equation labels as ordinary links -->
<div id="_auto1"></div>

$$
\begin{equation}
\mathbb{P}(Z_1=i,Z_3=j\vert Z_2=k) = \frac{\mathbb{P}(Z_1=i,Z_2=k,Z_3=j)}{\mathbb{P}(Z_2 =k)}
\label{_auto1} \tag{1}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="eq:condIndep"></div>

$$
\begin{equation} \
=\frac{\mathbb{P}(X_1 =i)\mathbb{P}(X_2 =k-i)\mathbb{P}(X_3 =j-k) }{\mathbb{P}(Z_2 =k)}
\end{equation} 
\label{eq:condIndep} \tag{2}
$$

onde a fatoração vem da independência de
as variáveis ​​$X_i$. **Observe que para $Z_1=i$, devemos ter $X_1=i$. Além disso, para $Z_2=X_1+X_2=k$ devemos ter $X_2 = k-i$ e assim por diante...**

Observe ainda que $\mathbb{P}(X_1 =i)\mathbb{P}(X_2 =k-i) = \mathbb{P}(Z_1 =i\vert Z_2=k) \mathbb{P}(Z_2=k) $, de modo que


$$
\mathbb{P}(Z_1=i,Z_3=j\vert Z_2=k) = \frac{\mathbb{P}(Z_1 =i\vert Z_2=k) \mathbb{P}(Z_2=k)\mathbb {P}(X_3 =j-k)}{\mathbb{P}(Z_2 =k)}
$$

Além disso, $\mathbb{P}(X_3 =j-k) = \mathbb{P}(Z_3 =j\vert Z_2=k)$ e, portanto:

$$
\mathbb{P}(Z_1=i,Z_3=j\vert Z_2=k) = \mathbb{P}(Z_1 =i\vert Z_2)\mathbb{P}(Z_3 =j\vert Z_2)
$$

**Assim, vemos que a dependência entre
variáveis ​​aleatórias podem ser quebradas por condicionamento para criar
variáveis ​​aleatórias condicionalmente independentes.**

Como acabamos de
testemunhado, compreendendo como o condicionamento influencia a independência
é importante e é o tema principal
estudo em Modelos Gráficos Probabilísticos, um campo
com muitos algoritmos e conceitos para extrair esses
noções de independência condicional de gráficos baseados
representações de variáveis ​​aleatórias.


## Exemplo Clássico de Haste Quebrada

Vamos fazer um último exemplo para exercitar a fluência em nossos métodos,
considerando o seguinte problema clássico: dada uma haste de comprimento unitário,
quebrado independente e aleatoriamente em dois lugares, qual é o
probabilidade de que você possa montar as três peças restantes em uma
triângulo? A primeira tarefa é encontrar uma representação de um triângulo como
uma restrição fácil de aplicar. O que queremos é algo como o
Segue:

$$
\mathbb{P}(\mbox{ triangle exists }) = \int_0^1 \int_0^1 \lbrace \mbox{ triangle exists }  \rbrace dX dY
$$

onde $X$ e $Y$ são independentes e uniformemente distribuídos
no intervalo unitário. A fórmula de Heron para a área do triângulo,

$$
\mbox{ area } = \sqrt{(s-a)(s-b)(s-c)s}
$$

onde $s = (a+b+c)/2$ é o que precisamos. **A ideia é que este
produz uma área válida somente quando cada um dos termos sob a raiz quadrada é
maior ou igual a zero.** Assim, suponha que temos

$$
\begin{eqnarray*}
a  & = &  X  \\\
b  & = & Y-X \\\
c  & = & 1-Y 
\end{eqnarray*}
$$

assumindo que $Y>X$. Assim, o critério para um triângulo válido resume-se
para

$$
\lbrace (s > a) \wedge (s > b) \wedge (s > c) \wedge (X<Y) \rbrace
$$

onde $\wedge = \&$ significa "e".

Depois de um pouco de manipulação, isso se consolida em:

$$
\Big\lbrace \frac{1}{2} < Y < 1 \bigwedge \frac{1}{2}(2 Y-1) < X < \frac{1}{2} \Big\rbrace
$$

Que integramos por $dX$ primeiro para obter

$$
\mathbb{P}(\mbox{ triangle exists }) = \int_{0}^1 \int_{0}^1 \Big\lbrace \frac{1}{2} < Y < 1 \bigwedge \frac{1}{2}(2 Y-1) < X < \frac{1}{2} \Big\rbrace dX dY
$$

$$
\mathbb{P}(\mbox{ triangle exists }) = \int_{\frac{1}{2}}^1 (1-Y) dY
$$

e então por $dY$ para obter finalmente,

$$
\mathbb{P}(\mbox{ triangle exists }) = \frac{1}{8}
$$

quando $Y>X$. Por simetria, obtemos o mesmo resultado para $X>Y$. Assim, o
resultado final é o seguinte:


$$
\mathbb{P}(\mbox{ triângulo existe }) = \frac{1}{8}+\frac{1}{8} = \frac{1}{4}
$$

Podemos verificar rapidamente usando este resultado usando Python para o caso $Y>X$ usando
o seguinte código:

```python
np.random.rand(2,1000)
```

```python
x,y = np.random.rand(2,100000)

a,b,c = y,(x-y),1-x # 3 sides

s = (a+b+c)/2

np.mean((s>a) & (s>b)  & (s>c) & (y<x))
```

# Métodos de projeção

O conceito de projeção é a chave para desenvolver uma intuição sobre condicional
probabilidade. Já temos uma intuição natural de projeção ao olhar para
as sombras dos objetos em um dia ensolarado. Como veremos, essa ideia simples consolida muitas ideias abstratas em otimização e matemática.

Considerar
[Figura](#fig:probabilidade_001) onde queremos encontrar um ponto ao longo do azul
linha (ou seja, $\mathbf{x}$) que está mais próxima do quadrado preto (ou seja,
$\mathbf{y}$). Em outras palavras, queremos inflar o círculo cinza até que apenas
toca a linha preta. Lembre-se de que o limite do círculo é o conjunto de pontos para
que

$$
\sqrt{(\mathbf{y}-\mathbf{x})^T(\mathbf{y}-\mathbf{x})} =\|\mathbf{y}-\mathbf{x} \| = \epsilon
$$

 por algum valor de $\epsilon$. Então, queremos um ponto $\mathbf{x}$ ao longo
a linha que satisfaz isso para o menor $\epsilon$. Então, esse ponto
será o ponto mais próximo da linha preta ao quadrado preto.
 Pode ser óbvio no diagrama, mas **o ponto mais próximo na linha
ocorre onde o segmento de linha do quadrado preto para a linha preta é
perpendicular à linha.** Neste ponto, o círculo cinza apenas toca o preto
linha. Isso é ilustrado abaixo:


**Dica de programação.**

[Figura](#fig:probabilidade_001) usa o módulo `matplotlib.patches`. este
módulo contém formas primitivas como círculos, elipses e retângulos que
podem ser montados em gráficos complexos. Conforme mostrado no código no IPython
Notebook correspondente a este capítulo, depois de importar uma determinada forma, você
pode aplicar essa forma a um eixo existente usando o método `add_patch`. o
os próprios patches podem ser estilizados usando as palavras-chave de formatação usuais, como
`cor` e `alfa`.

<p>O ponto mais próximo na linha ocorre quando a linha é tangente ao círculo. Quando isso acontece, a linha preta e a linha (distância mínima) são perpendiculares.
 

    
 Agora que podemos ver o que está acontecendo, podemos construir a solução
analiticamente. Podemos representar um ponto arbitrário ao longo da linha preta como:

$$
\mathbf{x}=\alpha\mathbf{v}
$$

onde $\alpha\in\mathbb{R}$ desliza o ponto para cima e para baixo na linha com

$$
\mathbf{v} = \left[ 1,1 \right]^T
$$

Formalmente, $\mathbf{v}$ é o *subespace* no qual queremos
*project* $\mathbf{y}$.

No ponto mais próximo, o vetor entre
$\mathbf{y}$ e $\mathbf{x}$ (o vetor de *error* acima) é
perpendicular à linha. Isso significa que

$$
(\mathbf{y}-\mathbf{x} )^T \mathbf{v} = 0
$$

e substituindo e resolvendo os termos, obtemos

$$
\alpha = \frac{\mathbf{y}^T\mathbf{v}}{ \|\mathbf{v} \|^2}
$$

Em notação vetorial temos:

$$
(\vec{y}-\vec{x})\cdot\vec{v} = 0 \\
\vec{y}\cdot\vec{v} = \vec{x}\cdot\vec{v}\\
\vec{y}\cdot\vec{v} = \alpha \vec{v}\cdot\vec{v}\\
\alpha = \frac{\vec{y}\cdot\vec{v}}{\vec{v}\cdot\vec{v}}
$$

O *erro* é a distância entre $\alpha\mathbf{v}$ e $
\mathbf{y}$.

Este é um triângulo retângulo, e podemos usar a fórmula pitagórica
teorema para calcular o comprimento ao quadrado deste erro como

$$
\epsilon^2 = \|( \mathbf{y}-\mathbf{x} )\|^2 = \|\mathbf{y}\|^2 - \alpha^2 \|\mathbf{v}\|^2 = \|\mathbf{y}\|^2 - \frac{\|\mathbf{y}^T\mathbf{v}\|^2}{\|\mathbf{v}\|^2}
$$

onde $ \|\mathbf{v}\|^2 = \mathbf{v}^T \mathbf{v} $. Observe que, como $\epsilon^2 \ge 0 $, isso também mostra que

$$
\| \mathbf{y}^T\mathbf{v}\| \le \|\mathbf{y}\|  \|\mathbf{v}\|
$$

que é a **famosa e útil desigualdade de Cauchy-Schwarz que
irá explorar mais tarde.**

Finalmente, podemos juntar tudo isso na *projeção*
operador

$$
\mathbf{P}_v = \frac{1}{\|\mathbf{v}\|^2 } \mathbf{v v}^T
$$

Em notação vetorial teríamos:

$$
\mathbf{P}_v = \frac{\vec{v}}{\vec{v}\cdot\vec{v}} \vec{v}\cdot
$$

Com este operador, podemos pegar qualquer $\mathbf{y}$ e encontrar o mais próximo
apontar em $\mathbf{v}$ fazendo

$$
\mathbf{P}_v \mathbf{y} = \mathbf{v} \left( \frac{  \mathbf{v}^T \mathbf{y} }{\|\mathbf{v}\|^2} \right)
$$

onde reconhecemos o termo entre parênteses como o $\alpha$ que
calculado anteriormente. É chamado de *operador* porque pega um vetor
($\mathbf{y}$) e produz outro vetor ($\alpha\mathbf{v}$). **Desta forma,
a projeção unifica geometria e otimização.**

## Distância ponderada

Podemos facilmente estender este operador de projeção para casos onde a medida de
distância entre $\mathbf{y}$ e o subespaço $\mathbf{v}$ é ponderada. Podemos
acomodar essas distâncias ponderadas reescrevendo o operador de projeção como

<!-- Equation labels as ordinary links -->
<div id="eq:weightedProj"></div>

$$
\begin{equation}
\mathbf{P}_v=\mathbf{v}\frac{\mathbf{v}^T\mathbf{Q}^T}{\mathbf{v}^T\mathbf{Q v}}
\end{equation}
\label{eq:weightedProj} \tag{1}
$$

onde $\mathbf{Q}$ é uma matriz definida positiva. No anterior
caso, partimos de um ponto $\mathbf{y}$ e inflamos um círculo centrado em
$\mathbf{y}$ até tocar a linha definida por $\mathbf{v}$ e isso
point era o ponto mais próximo na linha para $\mathbf{y}$. A mesma coisa acontece
no caso geral com uma distância ponderada, exceto que agora inflamos um
elipse, não um círculo, até que a elipse toque a linha.

<p>No caso ponderado, o ponto mais próximo da linha é tangente à elipse e ainda é perpendicular no sentido da distância ponderada.</p>



Observe que o vetor de erro ($\mathbf{y}-\alpha\mathbf{v}$) na [Figura] ainda é perpendicular à linha (subespaço
$\mathbf{v}$), mas no espaço da distância ponderada, ou seja,

$$(\mathbf{y}-\mathbf{x} )^T \mathbf{Q}^T\mathbf{v} = 0$$


A diferença
entre a primeira projeção (com a distância circular uniforme) e a
caso geral (com a distância elíptica ponderada) é o produto interno
entre os dois casos. Por exemplo, no primeiro caso temos $\mathbf{y}^T
\mathbf{v}$ e no caso ponderado temos $\mathbf{y}^T \mathbf{Q}^T
\mathbf{v}$. Para passar da caixa circular uniforme para a elipsoidal ponderada
caso, tudo o que tínhamos que fazer era mudar todos os produtos internos do vetor. Antes de nós
terminar, precisamos de uma propriedade formal de projeções:

$$
\mathbf{P}_v \mathbf{P}_v = \mathbf{P}_v
$$

conhecida como a propriedade *idempotente* que basicamente diz que uma vez que
projetamos em um subespaço, as projeções subsequentes nos deixam no
mesmo subespaço. Você pode verificar isso usando a definição

$$
\mathbf{P}_v=\mathbf{v}\frac{\mathbf{v}^T\mathbf{Q}^T}{\mathbf{v}^T\mathbf{Q v}}
$$
isso é,

$$
\mathbf{P}_v \mathbf{P}_v = \mathbf{v}\frac{\mathbf{v}^T\mathbf{Q}^T}{\mathbf{v}^T\mathbf{Q v} }\mathbf{v}\frac{\mathbf{v}^T\mathbf{Q}^T}{\mathbf{v}^T\mathbf{Q v}} = \mathbf{P}_v
$$


**Assim, a projeção vincula um problema de minimização (ponto mais próximo de uma linha) a um
conceito algébrico (produto interno).**

Acontece que essas mesmas ideias geométricas
da álgebra linear [[strang2006linear]](#strang2006linear) pode ser traduzido para o condicional
expectativa. Como isso funciona é o assunto de nossa próxima seção.

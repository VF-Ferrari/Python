## Matplotlib: plotagem

Matplotlib é provavelmente o pacote Python mais usado para gráficos 2D. Ele fornece uma maneira rápida de visualizar dados do Python e figuras com qualidade de publicação em vários formatos. Vamos explorar o matplotlib no modo interativo, cobrindo os casos mais comuns.

### pyplot

pyplot fornece uma interface procedural para a biblioteca de plotagem orientada a objeto matplotlib


#### Definindo marcações

````python
plt.plot(x,np.cos(x))
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1,0,1])
````

#### Definindo rótulos de marcação

````python
plt.plot(x,np.cos(x))
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
          ['$0$','$\pi/2$',r'$\pi$','$3\pi/2$','$2\pi$'])
plt.yticks([-1,0,1],['menos um','zero','um'])
````

#### Adicinando Legenda

````python
plt.plot(x,np.cos(x),label=r'$f(x)=cos(x)$')
plt.plot(x,np.sin(x),label=r'$f(x)=sin(x)$')
plt.legend()
#plt.legend(loc='upper left')
````


#### Anotar alguns pontos

````python
plt.plot(x,np.cos(x),label=r'$f(x)=cos(x)$')
plt.plot(x,np.sin(x),label=r'$f(x)=sin(x)$')
plt.legend()
t=np.pi/2
plt.annotate(r'$sin\left(\frac{\pi}{2}\right)=1$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(60, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=1"))
````

### Anatomia de uma figura
#### Figuras, Subtramas, Eixos e Marcas

````python
f, ax = plt.subplots(figsize=(5,3))
ax.plot(x,np.sin(x))
````

````python
f, ax = plt.subplots(ncols=2,nrows=1,figsize=(10,3))
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
````

````python
len(np.linspace(1,2,num=9))
````

````python
f, axs = plt.subplots(ncols=3,nrows=3,figsize=(10,10))

x = np.linspace(0,2*np.pi)
for omega,ax in zip(np.linspace(1,2,num=9),axs.flatten()):
    ax.plot(x,np.cos(omega*x))
````

````python
f, axs = plt.subplots(ncols=3,nrows=3,
                      figsize=(10,10),sharex=True,
                      sharey=True)

x = np.linspace(0,2*np.pi)
for omega,ax in zip(np.linspace(1,2,num=9),axs.flatten()):
    ax.plot(x,np.cos(omega*x))
````

## Medidas Experimentais

```python
a=10.0**-6
D=46.7

cores={'violeta1':0, 'violeta2':1,'verde':2,'amarelo':3}

xd=np.array([np.nan,22.8,29.8,32.2])
xe=np.array([20.7,22.6,29.3,31.5])
x=0.5*(xd+xe)

#Raia visível somente à esquerda
x[cores['violeta1'] ]=20.7
x
```

## Cores das Raias de Emissão

### $d\sin(\theta)=m\lambda \rightarrow \lambda=a\dfrac{x}{\sqrt{x^2+D^2}}$

```python
lambdas=a*x/np.sqrt(x**2+D**2)
```

```python
lambdas
```

## Qual a cor de um comprimento de onda?


```python
def cor(lamb):
    if lamb<380e-9:
        return 'Ultravioleta'
    elif lamb<450e-9:
        return 'Violeta'
    elif lamb<495e-9:
        return 'Azul'
    elif lamb<570e-9:
        return 'Verde'
    elif lamb<590e-9:
        return 'Amarelo'
    elif lamb<620e-9:
        return 'Laranja'
    elif lamb<750e-9:
        return 'Vermelho'
    else:
        return 'Infravermelho'
```

```python
cores_raias=[cor(lamb) for lamb in lambdas]
lambdas*1e9
```
